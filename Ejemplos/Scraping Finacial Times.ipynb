# Scraping Finacial Times

```python
import pandas as pd
```

```python
def risk(isin: str):
    url_base = 'https://markets.ft.com/data/funds/tearsheet/'
    currency='USD'

    periods = {0: '1 year', 1: '3 years', 2: '5 years'}
    url = f'{url_base}risk?s={isin}:{currency}'

    tables = pd.read_html(url, index_col=0)

    table_df = {}
    for num, name in periods.items():
        table_df[name] = pd.concat([tables[num], tables[num+1]])
    return  pd.concat(table_df, 1)   
```

```python
def performance(isin: str):
    url_base = 'https://markets.ft.com/data/funds/tearsheet/'
    currency = 'USD'

    url = f'{url_base}performance?s={isin}:{currency}'

    return pd.read_html(url, index_col=0)[0]
```

```python
def historical(isin: str):
    url_base = 'https://markets.ft.com/data/funds/tearsheet/'
    currency = 'USD'

    url = f'{url_base}historical?s={isin}:{currency}'

    historical = pd.read_html(url, index_col=0)[0]
    historical.index = pd.to_datetime(historical.index.str.split(', ').str[-2:].str.join(' '))
    return historical
```
